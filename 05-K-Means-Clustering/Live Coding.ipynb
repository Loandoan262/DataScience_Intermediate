# Elbow Method

```python
from scipy.spatial.distance import cdist
```

```python
K = 15
losses = []
for i in range (1,K):
    kmean_i = KMeans(n_clusters=i, random_state=0)
    kmean_i.fit(df)
    d2center = cdist(df, kmean_i.cluster_centers_)
    #print("i=", i,  "distance", d2center)
    min_d = np.min(d2center, axis=1)
    #print("i",i, "min", min_d)
    loss = np.sum(min_d)
    #print("loss", loss)
    losses.append(loss)
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.arange(1, K), losses, "-o")
```
