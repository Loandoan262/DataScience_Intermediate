# Text Pre-processing

```python
import string
```

```python
mess = "Hi, Please pay attention: No smoking!."
```

```python
nopunc = [char for char in mess if char not in string.punctuation]
nopunc = "".join(nopunc)
nopunc
```

```python
nopunc.split()
```

```python
[word for word in nopunc.split() if word.lower() not in stopwords.words("english")]
```

```python
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download("stopwords")
stopwords.words("english")
```

```python
def text_process(mess):
    """
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = "".join(nopunc)
    return [word.lower() for word in nopunc.split() if word.lower() not in stopwords.words("english")]
```

```python
messages
```

```python
messages["message"].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_trans = CountVectorizer(analyzer=text_process). fit(messages["message"])
```

```python
bow_trans.vocabulary_
```

```python
len(bow_trans.vocabulary_)
```

```python
message_test = messages["message"][3]
```

```python
message_test
```

```python
bow_message_test = bow_trans.transform([message_test])
print(bow_message_test)
```

```python
bow_trans.get_feature_names()[7208]
```

```python
bow_message_test.shape
```

```python
messages_bow = bow_trans.transform(messages["message"])
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer().fit(messages_bow)
tfidf_transformer
```

```python
tfidf_test = tfidf_transformer.transform(bow_message_test)
print(tfidf_test)
```

```python
messages
```

```python
messages_tiidf = tfidf_transformer.transform(messages_bow)
```

# Train model

```python
from sklearn.naive_bayes import MultinomialNB
```
