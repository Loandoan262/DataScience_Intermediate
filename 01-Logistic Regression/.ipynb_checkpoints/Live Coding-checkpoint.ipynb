```python
titanic["Age"]=titanic[["Age", "Pclass"]].groupby("Pclass").transform(lambda x: x.fillna(x.mean()).round())["Age"]
```

```python
titanic
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 38
        if Pclass == 2:
            return 30
        if Pclass == 3:
            return 25
    else:
        return Age
```

```python
titanic["Age"] = titanic[["Age", "Pclass"]].apply(impute_age, axis = 1)
```

```python
titanic
```

```python
sns.heatmap(titanic.isnull(), yticklabels=False)
```

```python
titanic.drop("Cabin", axis = 1, inplace = True)
```

```python
titanic
```

```python
# titanic.drop("Cabin", axis = 1, inplace = True) ========== titanic = titanic.drop("Cabin", axis = 1)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic
```

## Converting categorical features

```python
titanic.info()
```

```python
sex = pd.get_dummies(titanic["Sex"], drop_first=True)
embarked = pd.get_dummies(titanic["Embarked"], drop_first=True)
```

```python
titanic.drop(["Sex", "Embarked", "Name", "Ticket"], axis = 1 , inplace=True)
```

```python
titanic
```

```python
titanic_new = pd.concat([titanic, sex, embarked], axis = 1)
titanic_new
```

```python
titanic_copy = titanic_new.copy()
```

```python
titanic_copy["female"] = np.where(titanic_new["male"]==True, False, True)
```

```python
titanic_copy
```

# Building a model

```python
from sklearn.model_selection import train_test_split
```

```python
X = titanic_new.drop(["Survived"], axis = 1)
y = titanic_new["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=101)
```

## Training and Predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
pred = logmodel.predict(X_test)
pred
```

## Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```
