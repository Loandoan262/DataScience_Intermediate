# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaler_features, df["TARGET CLASS"], test_size=0.3, random_state=0)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

# Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(pred, y_test))
```

```python
#print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
def k_optimization(k, X, y, n):
    knn = KNeighborsClassifier(n_neighbors=k)
    kfold = KFold(n_splits=n, shuffle=True, random_state=0)
    score = cross_val_score(estimator=knn, X=X, y=y, cv=kfold, scoring="accuracy").mean()
    return score
```

```python
k_table = pd.DataFrame(columns = ["k", "score"])
for i in range(1, 40):
    #print("i: ", i)
    score = k_optimization(k=i, X=X_train, y=y_train, n=4)
    #print("i: ", i, score)
    k_table = pd.concat([k_table, pd.DataFrame({"k": [i], "score": [score]})], ignore_index=True)
k_table
```

```python
optimal_k = k_table["k"][k_table["score"] == k_table["score"].max()].iloc[0]
optimal_k
```

```python
knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(confusion_matrix(y_pred, y_test))
print(classification_report(y_test, pred))
```
