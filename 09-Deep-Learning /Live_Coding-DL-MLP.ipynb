```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
iris_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['species'])
iris_data.head()
```

```python
iris_data.info()
```

```python
X = iris_data.loc[:, iris_data.columns != "species"]
y = iris_data.loc[:, ["species"]]
```

```python
#y_enc = LabelEncoder().fit_transform(y)
y_label = tf.keras.utils.to_categorical(y)
```

```python
y_label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_label, test_size = 0.2, random_state=100)
```

```python
X_train.shape[1:]
```

```python
def get_model():
    model = Sequential([
        keras.layers.Input(shape = X_train.shape[1:]),
        keras.layers.Dense(800, activation="relu"),
        keras.layers.Dense(400, activation="relu"),
        keras.layers.Dense(200, activation="relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(3, activation="softmax")
    ])
    return model
```

```python
model = get_model()
```

```python
model.compile(optimizer="adam", 
              loss = keras.losses.CategoricalCrossentropy(),
              metrics=["accuracy"]
             )
```

```python
model.summary()
```

```python
fit_model = model.fit(X_train, y_train, epochs = 30, validation_data=(X_test, y_test), verbose = 1)
```
