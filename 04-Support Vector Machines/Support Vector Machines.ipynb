```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df_feat = pd.DataFrame(cancer["data"], columns = cancer["feature_names"])
df_feat
```

```python
df_target = pd.DataFrame(cancer["target"], columns = ["Cancer"])
```

```python
df_target.value_counts()
```

```python
df_feat.corr()
```

```python
df_target.shape
```

np.ravel(df_target).shape

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, np.ravel(df_target), test_size=0.3, random_state=101)
```

# Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
```

# Gridsearch

```python
para_grid = {"C": [0.1, 1, 10, 100, 1000], "gamma": [1.0, 0.1, 0.1, 0.001, 0.00001], "kernel": ["rbf"]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), para_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_pred))
```
